```python
from __future__ import absolute_import, division, print_function
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras
CUDA_VISIBLE_DEVICES=1 
# Helper libraries
import numpy as np
import cv2
import csv  
import os
from sklearn.model_selection import train_test_split
```

```python
gpu = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpu[0], True)
```

```python
def dataset(dir):
    letterLabel = []
    images = []
    for index, label in enumerate(uniqueLetterL):
        for file in os.listdir(dir + '/' + label):
            filepath = dir +'/'+ label + "/" + file
            image = cv2.resize(cv2.imread(filepath),(50,50))
            letterLabel.append(index)
            images.append(image)
        images = np.asarray(images)
        letterLabel = np.asarray(letterLabel)
        return images, letterLabel
    print(images)

#loading all the dataset
data_directory = r'C:\Users\Jason\Desktop\signLanguage\letters'
uniqueLetterL = sorted(os.listdir(data_directory))
X_pre, Y_pre = dataset(data_directory) 

dataset(data_directory)

```

```python

Y_train = tf.keras.utils.to_categorical(Y_pre)
X_train = X_pre / 255.




model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(16, (3,3), activation ='relu', input_shape=(50,50,3)),
        tf.keras.layers.Conv2D(16, (3,3), activation ='relu'),
        tf.keras.layers.Conv2D(16, (3,3), activation ='relu'),
        tf.keras.layers.MaxPool2D((2,2)),
        tf.keras.layers.Conv2D(32, (3,3), activation ='relu'),
        tf.keras.layers.Conv2D(32, (3,3), activation ='relu'),
        tf.keras.layers.Conv2D(32, (3,3), activation ='relu'),
        tf.keras.layers.MaxPool2D((2,2)),
        tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
        tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
        tf.keras.layers.Conv2D(64, (3,3), activation ='relu'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(25, activation='softmax')
])

model.summary()
```

```python
model.compile(optimizer = 'adam',
               loss = 'sparse_categorical_crossentropy',
               metrics=['accuracy'],)
```

```python
history = model.fit(X_train, Y_train, epochs=20, verbose = 1)
```
